```python
# Import
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as func
import torch.optim as opt

# 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = np.loadtxt(
    '/content/drive/MyDrive/Colab Notebooks/code_datacsv1.csv', # 드라이브 마운트
    delimiter=',',
    dtype=np.float32)
```

```python
x_t = torch.FloatTensor(dataset[:,:-1])
y_t = torch.LongTensor(dataset[:,-1]) #cross entropy를 쓰려면 출력값을 Long 으로 받아야하는 규칙이 있다고 한다.
model_e = nn.Linear(6,3)
```

```python
print(x_t)
print(x_t.shape)
print(y_t)
print(y_t.shape)
```

```python
# Optimizer 설정 (learning rate = 0.0000004로 설정)
optimizer_e = opt.SGD(model_e.parameters(), lr=0.0000004)

# 반복
for epoch in range(500000):

  # Cost 계산
  y_h = model_e(x_t)
  cost_e = func.cross_entropy(y_h, y_t)

  # cost를 이용해 model update
  optimizer_e.zero_grad()
  cost_e.backward()
  optimizer_e.step()

  # 100번 마다 중간결과 출력
  if epoch % 100 == 99:
    correct_prediction = y_h.argmax(dim=1) == y_t
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    print('Epoch {:7d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, 500000, cost_e.item(), accuracy * 100,
    ))
```
